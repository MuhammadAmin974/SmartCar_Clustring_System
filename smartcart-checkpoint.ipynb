```python
Ohe=OneHotEncoder(sparse_output=False,handle_unknown="ignore")
Encoded=Ohe.fit_transform(df[["Education","Living_with"]])
```

```python
Encoded_df=pd.DataFrame(Encoded,columns=Ohe.get_feature_names_out(["Education","Living_with"]))
Encoded_df
```

```python
df=df.drop(["Education","Living_with"],axis=1)
```

```python
df=pd.concat([Encoded_df,df],axis=1)
```

```python
df.isnull().sum()
```

```python
scale=StandardScaler()
Scaled_df=pd.DataFrame(scale.fit_transform(df),columns=df.columns)
```

```python
Scaled_df.head()
```

```python
Scaled_df=Scaled_df.dropna()
```

```python
Scaled_df.isnull().sum()
```

```python
Scaled_df.columns
```

```python
Scaled_df.shape
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=3)

pca_df=pca.fit_transform(Scaled_df)
plt.scatter(pca_df[1:,0],pca_df[1:,1])
pca.explained_variance_
```

```python
from kneed import KneeLocator
from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering

from sklearn.pipeline import make_pipeline
```

```python
wcss=[]
for k in range(1,11):
    kmeans=KMeans(n_clusters=k,random_state=42)
    kmeans.fit_predict(pca_df)
    wcss.append(kmeans.inertia_)

kl=KneeLocator(range(1,11),wcss,curve="convex",direction="decreasing")

print(kl.elbow)
```

```python
kmeans=KMeans(n_clusters=7,random_state=42)
agg=AgglomerativeClustering(n_clusters=7,linkage="ward")
```

```python
kmeans.fit(pca_df)
```

```python
from sklearn.metrics import silhouette_score
```

```python
score=silhouette_score(pca_df,kmeans.labels_)
print(score)
```

```python
agg.fit(pca_df)
```

```python
score=silhouette_score(pca_df,agg.labels_)
print(score)
```
